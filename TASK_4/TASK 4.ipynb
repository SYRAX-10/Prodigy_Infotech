```python
#get dummy 

dummy = pd.get_dummies(df2[['Age_band_of_driver','Vehicle_driver_relation', 'Driving_experience','Area_accident_occured', 'Lanes_or_Medians','Types_of_Junction', 'Road_surface_type','Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement', 'Casualty_class','Age_band_of_casualty','Cause_of_accident']])
dummy.head()
```

```python
#concatinate dummy and old data frame 

df3 = pd.concat([df2,dummy],axis=1)
df3.head()
```

```python
#seperating independent and dependent 

df3.drop(['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience', 'Area_accident_occured', 'Lanes_or_Medians','Types_of_Junction', 'Road_surface_type', 'Light_conditions', 'Weather_conditions', 'Type_of_collision',
'Vehicle_movement','Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident'], axis=1, inplace=True)

df3.head()
```

```python
x = df3.drop(["Accident_severity"],axis=1)
x.shape
```

```python
y= df3.iloc[:,4]
y.head()
```

```python
y.value_counts()
```

```python
#plotting count plot using seaborn

sns.countplot(x=y, palette = 'muted')
```

```python
# Convert 'Time' column to numerical format
x['Time'] = pd.to_datetime(x['Time']).dt.hour + pd.to_datetime(x['Time']).dt.minute / 60

# Convert categorical columns to numerical values
categorical_columns = x.select_dtypes(include=['object']).columns
x[categorical_columns] = x[categorical_columns].apply(lambda col: le.fit_transform(col))

# Oversampling 
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
xo, yo = oversample.fit_resample(x, y)
```

```python
#checking the oversampling output 

y1 = pd.DataFrame(yo)
y1.value_counts()
```

```python
sns.countplot(x =yo, palette = 'muted')
```

```python
#splitting the data

from sklearn.model_selection import train_test_split

#splitting 70% of the data to training data and 30% of data to testing data 

x_train,x_test,y_train,y_test = train_test_split(xo,yo,test_size=0.30,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#KNN Model Creation

from sklearn.neighbors import KNeighborsClassifier

model_KNN= KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred = model_KNN.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
```

```python
report_KNN = classification_report(y_test,y_pred)
print(report_KNN)
```

```python
accuracy_KNN = accuracy_score(y_test,y_pred)
print(accuracy_KNN)
```

```python
#Confuion Matrix

matrix_KNN = confusion_matrix(y_test,y_pred)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
